```python
link = "https://ic2s2-2023.org/program"
req = requests.get(link)
soup = BeautifulSoup(req.content)
```

Extracting names with i meaning chairs, parallel talks and authors of posters

```python
table_rows = soup.find_all("i")
names = []
for line in table_rows:
    line = line.get_text(strip=True).replace(":", ",").split(",")
    if "Chair" in line:
        line.remove("Chair")
    names.extend(name.strip() for name in line if name.strip())
names = list(dict.fromkeys(names))

print((names))

print(len(names))
```

Adding keynames and removing duplicates

```python
# Find all <a> tags that mention "Keynote"
keynote_links = soup.find_all("a", string=lambda text: text and "Keynote" in text)

# Extract and clean names
keynote_speakers = []
for link in keynote_links:
    text = link.get_text(strip=True)  # Get text inside <a> tag
    name = text.replace("Keynote - ", "").strip()  # Remove "Keynote - "
    keynote_speakers.append(name)
# Print all keynote speakers
if "Keynotes" in keynote_speakers:
    keynote_speakers.remove("Keynotes")

print(keynote_speakers)
names.extend(keynote_speakers)

#Remove duplicates
names = list(set(names))
names = list({name.strip().lower() for name in names})
print(names)
print(len(names))
```

The following code is updated so instead of slicing it takes use of the get function from the openalex API.

It returns each IC2S2 author, their ID and a coloumn containing relevant work information not including abstract which comes in a different coloumn.

```python
import requests
import pandas as pd
from joblib import Parallel, delayed
from tqdm import tqdm
import time

def get_author(author_id):
    try:
        # Ensure author_id is a string
        author_id = str(author_id)

        # 1. Get basic author info
        author_url = f"https://api.openalex.org/authors/{author_id}"
        author_response = requests.get(author_url)
        if author_response.status_code != 200:
            return {"author_id": author_id, "error": "Author not found"}

        author_data = author_response.json()
        author_name = author_data.get("display_name", "Unknown")

        # 2. Get all author's works with pagination
        base_works_url = f"https://api.openalex.org/works?filter=author.id:{author_id}&per_page=200"
        all_works = []

        cursor = "*"
        while cursor:
            works_url = f"{base_works_url}&cursor={cursor}"
            works_response = requests.get(works_url)
            if works_response.status_code != 200:
                break
            works_data = works_response.json()
            works = works_data.get("results", [])
            all_works.extend(works)
            cursor = works_data.get("meta", {}).get("next_cursor", None)
            time.sleep(0.5)

        # Initialize data structure
        works_data = {
            "author_id": author_id,
            "author_name": author_name,
            "works": [],
            "abstract": []
        }

        # Process all collected works and check concepts
        for work in all_works:
            concepts = {c["display_name"] for c in work.get("concepts", []) if c["level"] == 0}
            social_science = {"Sociology", "Psychology", "Economics", "Political Science"}         
            quantitative_science = {"Mathematics", "Physics", "Computer Science"}
            if not (concepts & social_science) or not (concepts & quantitative_science):
                continue
            work_info = {
                "id": work.get("id"),
                "title": work.get("title"),
                "publication_year": work.get("publication_year"),
                "cited_by_count": work.get("cited_by_count"),
                "author_ids": [authorship.get("author", {}).get("id") 
                             for authorship in work.get("authorships", []) 
                             if authorship.get("author", {}).get("id")]
            }
            works_data["abstract"].append({"abstracts_inverted_index": work.get("abstract_inverted_index", "N/A")})
            works_data["works"].append(work_info)

        return works_data

    except Exception as e:
        return {"author_id": author_id, "error": str(e)}
```

We run the above function for all author IDs

```python
# Load and clean DataFrame
df = pd.read_csv("authors_matched.csv")
df['ID'] = df['ID'].str.split('/').str[-1]
df = df[df["Works Count"] > 5]
df = df[df["Works Count"] < 5000]

# Run in parallel with progress bar
all_extracted_data = Parallel(n_jobs=5)(
    delayed(get_author)(author) for author in tqdm(df["ID"])
)

# Convert results to DataFrame (optional)
results_df = pd.DataFrame(all_extracted_data)
```

```python
import networkx as nx
import ast
from networkx.readwrite import json_graph
import json
```

```python
papers_df = pd.read_csv("IC2S2_papers.csv")

# Convert string representation of lists into actual Python lists
papers_df["author_ids"] = papers_df["author_ids"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
papers_df
```

```python
authors_df = pd.read_csv("authors_matched.csv")
authors_df = authors_df.drop_duplicates(subset="ID")
authors_df
```

```python
from collections import defaultdict
from itertools import combinations

def create_edge_list(df, authors_column):
    collaborations = defaultdict(int)
    
    for authors in df[authors_column]:
        for author1, author2 in combinations(sorted(authors), 2):
            collaborations[(author1, author2)] += 1
    
    edge_list = [(author1, author2, count) for (author1, author2), count in collaborations.items()]
    return edge_list

edge_list = create_edge_list(papers_df,'author_ids')
edge_list
```

```python
def create_graph(G,edge_list):
    for author1, author2, _ in edge_list:
        G.add_edge(author1,author2)
    return G

def add_weights(G,edge_list):
    G.add_weighted_edges_from(edge_list)
    return G

G = nx.Graph()
G = create_graph(G,edge_list)
G = add_weights(G,edge_list)
print(G)
```

```python
def node_attributes(G, authors_df, papers_df):
    authors_info = authors_df.set_index('ID')[['Name', 'Country Code']].to_dict(orient='index')
    
    publication_info = {}

    for idx, row in papers_df.iterrows():
        author_list = row["author_ids"]
       
        pub_year = row['publication_year']
        cited_by = row['cited_by_count']
        
        for author in author_list:
            if author not in publication_info:
                publication_info[author] = {
                    'first_publication_year': pub_year,
                    'citation_count': cited_by
                }
            else:
                if pub_year < publication_info[author]['first_publication_year']:
                    publication_info[author]['first_publication_year'] = pub_year
                publication_info[author]['citation_count'] += cited_by
    
    for author in G.nodes():
        author_attr = authors_info.get(author, {})
        display_name = author_attr.get('Name')
        country = author_attr.get('Country Code')
        
        pub_info = publication_info.get(author, {})
        first_year = pub_info.get('first_publication_year')
        citation_count = pub_info.get('citation_count')
        
        combined_attr = {
            'display_name': display_name,
            'country': country,
            'first_publication_year': first_year,
            'citation_count': citation_count
        }
        G.nodes[author].update(combined_attr)
    
    return G
```

```python
G = node_attributes(G,authors_df,papers_df)
```

```python
def save_graph_as_json(G, filename):
    data = json_graph.node_link_data(G)

    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)
```

```python
save_graph_as_json(G,"graph.json")
```
